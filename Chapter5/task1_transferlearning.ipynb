# mlp模型迁移数据预测

```python
#数据加载
import pandas as pd
import numpy as np
data1 = pd.read_csv('task1_data1.csv')
data1.head()
```

```python
#X y赋值
X = data1.loc[:,'x']
y = data1.loc[:,'y']
y.head()
```

```python
#数据可视化
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.scatter(X,y)
plt.title('y vs x data1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#维度确认
print(X.shape,y.shape)
X = np.array(X).reshape(-1,1)
print(X.shape)
```

```python
#建立模型
from keras.models import Sequential
from keras.layers import Dense
#创建模型
model1 = Sequential()
#添加隐藏层
model1.add(Dense(units=50, input_dim=1, activation='relu'))
model1.add(Dense(units=50,activation='relu'))
#输出层
model1.add(Dense(units=1,activation='linear'))
model1.summary()
```

```python
#参数配置
model1.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#模型训练
model1.fit(X,y,epochs=500)
```

```python
#模型预测与表现评估
y_predict = model1.predict(X)
#r2 score
from sklearn.metrics import r2_score
r2 = r2_score(y,y_predict)
print(r2)
```

```python
fig2 = plt.figure()
plt.scatter(X,y,label='raw data')
plt.scatter(X,y_predict,label='predict result')
plt.title('y vs x data1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#模型训练 500+500=1000次
model1.fit(X,y,epochs=500)
```
